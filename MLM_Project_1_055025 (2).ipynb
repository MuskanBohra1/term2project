#Clustering Report

### **1. Introduction**  
This report summarizes the results of a cluster analysis conducted on a preprocessed dataset. The primary objective of the analysis was to identify distinct groups or segments within the data based on shared characteristics, which can provide actionable insights for decision-making.  

---

### **2. Methodology**  

#### **2.1 Clustering Techniques**  
Two clustering algorithms were employed:  
1. **K-Means Clustering**: A centroid-based method known for its efficiency in segmenting data into a predefined number of clusters.  
2. **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**: A density-based clustering method that identifies clusters of arbitrary shape and detects outliers.  

#### **2.2 Steps Followed**  
1. **Data Preprocessing**:  
   - Relevant variables were selected for clustering.  
   - The dataset was scaled using the **MinMaxScaler** to standardize the range of values between 0 and 1.  

2. **Clustering Process**:  
   - K-Means was tested with varying numbers of clusters (from 2 to 5).  
   - DBSCAN was implemented to determine clusters based on density thresholds.  

3. **Cluster Evaluation**:  
   - Two metrics were used to assess cluster quality:  
     - **Silhouette Score**: Measures how well each point is clustered based on cohesion and separation.  
     - **Davies-Bouldin Score**: Evaluates cluster compactness and separation, with lower values indicating better clustering.  

4. **Cluster Selection**:  
   - The optimal number of clusters was determined by balancing the Silhouette and Davies-Bouldin scores.  

5. **Cluster Characterization**:  
   - Each cluster's characteristics were analyzed through descriptive statistics and visualizations to provide meaningful interpretations.  

---

### **3. Results**  

#### **3.1 Optimal Number of Clusters (K-Means)**  
The following table presents the evaluation metrics for K-Means clustering with varying numbers of clusters:  

| **Number of Clusters** | **Silhouette Score** | **Davies-Bouldin Score** |  
|-------------------------|----------------------|--------------------------|  
| 2                       | 0.2465              | 1.6313                  |  
| 3                       | 0.2439              | 1.3125                  |  
| 4                       | 0.2641              | 1.1601                  |  
| 5                       | 0.2701              | 1.0655                  |  

- The **5-cluster model** achieved the highest Silhouette Score (0.2701) and the lowest Davies-Bouldin Score (1.0655), indicating the most compact and well-separated clusters among the tested configurations.  

#### **3.2 Optimal Number of Clusters (DBSCAN)**  
DBSCAN produced **3 clusters** based on the density parameters (`eps` and `min_samples`).  
- While DBSCAN effectively captured natural groupings in the data and identified outliers, its clusters were fewer and less distinct compared to K-Means.  

#### **3.3 Cluster Comparisons**  
- **K-Means** provided clearer segmentation and was better suited for datasets with pre-specified cluster numbers.  
- **DBSCAN** excelled in detecting outliers and creating clusters of varying densities. However, it struggled with datasets requiring compact clusters.  

---

### **4. Insights and Recommendations**  

1. **Preferred Model**:  
   - For structured clustering tasks, **K-Means with 5 clusters** is recommended due to its superior evaluation metrics.  
   - Use **DBSCAN** for anomaly detection or exploratory purposes where identifying noise or irregularities is critical.

```python
def categorical_analysis(df):
    results = []
    for column in df.columns:
        mode_values = df[column].mode()
        mode = mode_values.iloc[0] if not mode_values.empty else np.nan
        frequency = df[column].value_counts().get(mode, 0)



        row = {
            'Variable': column,
            'Count': df[column].count(),
            'Mode': mode,
            'Frequency': frequency,
        }

        # ... (calculations for ordinal variables) ...

        results.append(row)

    return pd.DataFrame(results)

# Apply the function to your DataFrame
analysis_results_df = categorical_analysis(mb25_ppd_cat)

# Display the table
display(analysis_results_df)
```

```python
def plot_categorical_data(df, top_n=10):
    categorical_cols = mb25_ppd_cat.columns  # Get all categorical columns

    for col in categorical_cols:
        # Count the occurrences of each category
        counts = df[col].value_counts()

        # Check the number of unique values to decide visualization type
        num_unique = len(counts)

        plt.figure(figsize=(10, 6))

        if num_unique <= 10:
            # Pie chart for categories with <= 10 unique values
            plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
            plt.title(f'Pie Chart of {col}')
            plt.axis('equal')  # Equal aspect ratio ensures the pie chart is a circle
        elif num_unique <= 20:
            # Horizontal bar chart for categories with > 10 and <= 20 unique values
            sns.barplot(x=counts.values, y=counts.index, orient='h')
            plt.title(f'Bar Chart of {col}')
            plt.xlabel('Frequency')
            plt.ylabel(col)
        else:
            # Top N categories for larger datasets (>20 unique values)
            top_counts = counts.head(top_n)
            sns.barplot(x=top_counts.values, y=top_counts.index, orient='h')
            plt.title(f'Top {top_n} Bar Chart of {col}')
            plt.xlabel('Frequency')
            plt.ylabel(col)

        plt.show()

# Example usage with your dataset 'mb25_ppd_cat'
plot_categorical_data(mb25_ppd_cat)
```

```python
# valid_labels_dbs = dbscan_model[dbscan_model != -1]
# valid_data_dbs = mb25_ppd_subset[dbscan_model != -1]

# # Calculate Silhouette Score for the entire dataset (excluding noise points)
# if len(valid_labels_dbs) > 1:  # Ensure at least 2 points remain
#     mb25_sscore_dbs = sscore(valid_data_dbs, valid_labels_dbs)
#     print(f'DBS Silhouette Score (excluding noise) = {mb25_sscore_dbs}')
# else:
#     print("Not enough data points to calculate Silhouette Score.")
# if len(valid_labels_dbs) > 1:  # Ensure at least 2 points remain
#     mb25_dbscore_dbs = dbscore(valid_data_dbs, valid_labels_dbs)
#     print(f'DBS DBScore (Davies-Bouldin Index) = {mb25_dbscore_dbs}')
# else:
#     print("Not enough data points to calculate DBScore.")

!pip install kneed
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score as sscore, davies_bouldin_score as dbscore
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import NearestNeighbors
from kneed import KneeLocator


# Function to find optimal eps for DBSCAN using k-distance graph
def find_optimal_eps(data, k=5):
    """
    Finds the optimal eps value for DBSCAN using a k-distance graph.

    Args:
      data (pd.DataFrame): The dataset for clustering.
      k (int): The number of nearest neighbors to use for the k-distance graph.

    Returns:
      float: The optimal eps value.
    """
    neighbors = NearestNeighbors(n_neighbors=k)
    neighbors_fit = neighbors.fit(data)
    distances, indices = neighbors_fit.kneighbors(data)
    distances = np.sort(distances, axis=0)
    distances = distances[:, k-1]
    plt.figure(figsize=(10, 6))
    plt.plot(distances)
    plt.xlabel('Points sorted by distance to {}-th nearest neighbor'.format(k))
    plt.ylabel('Distance to {}-th nearest neighbor'.format(k))
    plt.title('K-distance Graph')
    plt.grid(True)
    plt.show()

    # Heuristic to find the elbow point (you might need to adjust this)
    try:
      #Finds point of maximum curvature.
      kneedle = KneeLocator(x=range(0, len(distances)), y=distances, S=1.0, curve="convex", direction="increasing")
      eps_optimal = distances[kneedle.elbow]
      print("Optimal eps value using KneeLocator: {:.3f}".format(eps_optimal))
      return eps_optimal

    except Exception as e:
      print(f"Error finding optimal eps using kneedle {e}, using manual method")
      # Manual method
      try:
        max_curvature_index = np.argmax(np.diff(distances, n=2)) + 1
        eps_optimal = distances[max_curvature_index]
        print("Optimal eps value using manual method {:.3f}".format(eps_optimal))
        return eps_optimal

      except Exception as e:
        print(f"Error finding optimal eps using manual method {e}, using 0.5 as fallback")
        return 0.5 #Fallback Value.

# Find the optimal eps value using the k-distance graph
optimal_eps = find_optimal_eps(mb25_ppd_subset)
# DBSCAN clustering with tuned eps
dbscan_cluster = DBSCAN(eps=optimal_eps, min_samples=5, metric='euclidean')  # Adjust min_samples if needed
dbscan_model = dbscan_cluster.fit_predict(mb25_ppd_subset)

print("DBSCAN model:", dbscan_model)

# Count clusters (excluding noise labeled as -1)
n_clusters = len(set(dbscan_model)) - (1 if -1 in dbscan_model else 0)
n_noise = list(dbscan_model).count(-1)
print(f"Number of clusters: {n_clusters}, Noise points: {n_noise}")

valid_labels_dbs = dbscan_model[dbscan_model != -1]
valid_data_dbs = mb25_ppd_subset[dbscan_model != -1]

# Calculate Silhouette Score for the entire dataset (excluding noise points)
if len(set(valid_labels_dbs)) > 1:  # Ensure at least 2 points remain
    mb25_sscore_dbs = sscore(valid_data_dbs, valid_labels_dbs)
    print(f'DBS Silhouette Score (excluding noise) = {mb25_sscore_dbs}')
    mb25_dbscore_dbs = dbscore(valid_data_dbs, valid_labels_dbs)
    print(f'DBS DBScore (Davies-Bouldin Index) = {mb25_dbscore_dbs}')
else:
    print("Not enough data points to calculate Silhouette or Davies-Bouldin Score. DBSCAN may have found only noise or 1 cluster.")

mb25_ppd_subset_dbscan = mb25_ppd_subset.copy()
mb25_ppd_subset_dbscan.reset_index(inplace=True, drop=True)
mb25_ppd_subset_dbscan['cluster_number'] = dbscan_model

# Get unique cluster labels (excluding noise -1)
unique_labels = sorted(list(set(dbscan_model) - {-1}))

# Create color map with enough unique colors
colors = plt.cm.get_cmap('viridis', len(unique_labels))

# Plot clusters
plt.figure(figsize=(8, 6))
for i, label in enumerate(unique_labels):
    cluster_data = mb25_ppd_subset_dbscan[mb25_ppd_subset_dbscan['cluster_number'] == label]
    plt.scatter(cluster_data['Quantity'], cluster_data['Value'],
                color=colors(i), label=f'Cluster {label}', edgecolors='k')

# Plot noise points if any
noise_data = mb25_ppd_subset_dbscan[mb25_ppd_subset_dbscan['cluster_number'] == -1]
if len(noise_data) > 0:
    plt.scatter(noise_data['Quantity'], noise_data['Value'],
                color='red', label='Noise', edgecolors='k')

plt.title(f'DBSCAN Clustering (eps={dbscan_cluster.eps}, min_samples={dbscan_cluster.min_samples})')
plt.xlabel('Quantity')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# K-Means Diagnostics
def kmeans_diagnostics(data, model):
    labels = model.labels_
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)  # Correctly count clusters

    # Cluster characteristics
    cluster_characteristics = []
    for i in range(n_clusters):
      cluster_data = data[labels == i]
      # Calculate homogeneity (low standard deviation indicates high homogeneity)
      homogeneity = cluster_data.std().mean()
      cluster_characteristics.append(
          {"cluster_id": i, "homogeneity": homogeneity}
      )

    # Calculate heterogeneity (using distance between cluster centroids)
    centroids = model.cluster_centers_
    distances = np.linalg.norm(centroids[:, np.newaxis, :] - centroids, axis=2)  # Pairwise distance
    np.fill_diagonal(distances, np.inf) # Ignore self distance
    heterogeneity = np.min(distances)

    return pd.DataFrame(cluster_characteristics), heterogeneity


# Example usage for 5-cluster KMeans model:
cluster_characteristics, heterogeneity = kmeans_diagnostics(mb25_ppd_subset, mb25_km_5cluster)

print("Cluster Characteristics (Homogeneity):")
print(cluster_characteristics)

print("\nHeterogeneity (Min distance between centroids):", heterogeneity)

# Silhouette and Davies-Bouldin scores:
silhouette = sscore(mb25_ppd_subset, mb25_km_5cluster_model)
db_index = dbscore(mb25_ppd_subset, mb25_km_5cluster_model)

print(f"\nSilhouette Score: {silhouette}")
print(f"Davies-Bouldin Index: {db_index}")
```
